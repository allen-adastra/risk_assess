```python
from scipy.io import loadmat
import matplotlib.pyplot as plt

res = loadmat("scenario_13_comp_0_results.mat")
rb_order_4 = res['result_struct']['rb_order_4'][0][0][0]
rb_order_2 = res['result_struct']['rb_order_2'][0][0][0]
cheb = res['result_struct']['chebyshev'][0][0][0]
plt.plot(rb_order_4, linewidth = 4)
plt.plot(rb_order_2,'r', linewidth = 4)
plt.plot(cheb, '.g', linewidth = 4)
plt.xlabel("Time Step")
plt.ylabel("Risk Bound")
plt.legend(["Fourth Order SOS", "Second Order SOS", "Chebyshev Bound"])
```
